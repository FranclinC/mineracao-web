```python
import pandas as pd
import numpy as np
```

```python
USERS_PATH = "cresci-2015/TFP/users.csv"
TWEETS_PATH = "cresci-2015/TFP/tweets.csv"
PROCESSED_USERS_PATH = "cresci-2015/TFP/users_processed.csv"
```

```python
pd.set_option('display.max_columns', 500)

users_df = pd.read_csv(USERS_PATH)
tweets_df = pd.read_csv(TWEETS_PATH)
```

```python
users_df = users_df[users_df["id"].isin(tweets_df["user_id"].unique())]
users_df.head(n=10)
```

```python
tweets_df.head(n=10)
```

```python
users = users_df.loc[:, ["id", "verified", "followers_count", "favourites_count", "friends_count"]]
users["friend_follower_ratio"] = users["friends_count"] / users["followers_count"]
users.reset_index(drop=True, inplace=True)
users.shape
```

```python
users.head(n=10)
```

```python
users["mentions"] = tweets_df["num_mentions"].groupby(tweets_df["user_id"]).mean().reset_index(drop=True)
users["hashtags"] = tweets_df["num_hashtags"].groupby(tweets_df["user_id"]).mean().reset_index(drop=True)
users["urls"] = tweets_df["num_urls"].groupby(tweets_df["user_id"]).mean().reset_index(drop=True)
users["retweets"] = tweets_df["retweet_count"].groupby(tweets_df["user_id"]).mean().reset_index(drop=True)
users["favorites"] = tweets_df["favorite_count"].groupby(tweets_df["user_id"]).mean().reset_index(drop=True)
```

```python
users.head(n=10)
```

```python
for index, uid in enumerate(users["id"]):
    user_tweets = tweets_df[tweets_df["user_id"] == uid]
    unique_places = user_tweets["place"].value_counts()[user_tweets["place"].value_counts() == 1].shape[0]
    users.loc[index, "unique_places"] = unique_places
```

```python
users.head(n=10)
```

```python
for index, uid in enumerate(users["id"]):
    tweets_hour = []
    user_tweets = tweets_df[tweets_df["user_id"] == uid].sort_values(by="timestamp", ascending=False)
    user_tweets.reset_index(inplace=True, drop=True)
    S = pd.to_datetime(user_tweets["timestamp"])
    for key, values in user_tweets.groupby([(S - S[0]).astype("timedelta64[h]")]):
        tweets_hour.append(len(values))
        
    tweets_hour = np.array(tweets_hour)
    users.loc[index, "tweet_rate_variance"] = np.var(tweets_hour)
```

```python
users.head(n=10)
```

```python
users.to_csv(PROCESSED_USERS_PATH)
```
