```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
```

```python
human_files = [
    "cresci-2017/datasets_full/genuine_accounts/users_processed.csv",
    "cresci-2015/E13/users_processed.csv",
    "cresci-2015/TFP/users_processed.csv"
]
bot_files = [
    "cresci-2017/datasets_full/social_spambots_1/users_processed.csv",
    "cresci-2017/datasets_full/social_spambots_2/users_processed.csv",
    "cresci-2017/datasets_full/social_spambots_3/users_processed.csv",
    "cresci-2017/datasets_full/traditional_spambots_1/users_processed.csv",
]

human_dfs = []
for f in human_files:
    human_dfs.append(pd.read_csv(f))
human_df = pd.concat(human_dfs)

bot_dfs = []
for f in bot_files:
    bot_dfs.append(pd.read_csv(f))
bot_df = pd.concat(bot_dfs)
```

```python
bot_df["bot"] = 1
bot_df.head(n=10)
```

```python
human_df["bot"] = 0
human_df.head(n=10)
```

```python
dataset_df = pd.concat([human_df, bot_df])
dataset_df.reset_index(drop=True, inplace=True)
dataset_df.head(n=10)
```

```python
dataset_df = dataset_df.iloc[:, 1:]
dataset_df.head(n=10)
```

```python
dataset_df[dataset_df["verified"] == 1]
```

```python
dataset_df["verified"] = dataset_df["verified"].fillna(0)
dataset_df["friend_follower_ratio"] = dataset_df["friend_follower_ratio"].fillna(1)
dataset_df.replace(np.inf, 100, inplace=True)
dataset_df.head(n=10)
```

```python
X = np.array(dataset_df.iloc[:, 1:-1])
X = (X - X.mean(axis=0)) / X.std(axis=0)
y = dataset_df.iloc[:, -1]
```

```python
print(pd.value_counts(pd.Series(y), normalize=True))
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=1/5, random_state=42, stratify=y)
```

```python
print(pd.value_counts(pd.Series(y_train), normalize=True))
print(pd.value_counts(pd.Series(y_val), normalize=True))
```

```python
np.save("X_train.npy", X_train)
np.save("y_train.npy", y_train)
np.save("X_val.npy", X_val)
np.save("y_val.npy", y_val)
```
